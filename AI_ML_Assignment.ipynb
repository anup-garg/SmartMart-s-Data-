```python
Q1 = df['Last_Month_Spend'].quantile(0.25)
Q3 = df['Last_Month_Spend'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Last_Month_Spend'] < lower_bound) | (df['Last_Month_Spend'] > upper_bound)]
print(outliers)
```

```python
median = df['Last_Month_Spend'].median()
df['Last_Month_Spend'] = df['Last_Month_Spend'].apply(lambda x: median if (x < lower_bound or x > upper_bound) else x)
```

```python
Q1 = df['Monthly_Sales_Prediction'].quantile(0.25)
Q3 = df['Monthly_Sales_Prediction'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['Monthly_Sales_Prediction'] < lower_bound) | (df['Monthly_Sales_Prediction'] > upper_bound)]
print(outliers)
```

```python
median = df['Monthly_Sales_Prediction'].median()
df['Monthly_Sales_Prediction'] = df['Monthly_Sales_Prediction'].apply(lambda x: median if (x < lower_bound or x > upper_bound) else x)
```

3.	Is there a significant difference in Annual_Income between churned and non-churned customers?

```python
from scipy import stats

churned_income = df[df['Will_Churn'] == 1]['Annual_Income']
non_churned_income = df[df['Will_Churn'] == 0]['Annual_Income']

ttest_result = stats.ttest_ind(churned_income, non_churned_income)

print(f"T-test statistic: {ttest_result.statistic}")
print(f"P-value: {ttest_result.pvalue}")

alpha = 0.05
if ttest_result.pvalue < alpha:
    print("There is a significant difference in Annual_Income between churned and non-churned customers.")
else:
    print("There is no significant difference in Annual_Income between churned and non-churned customers.")
```

8.	Does the existing Monthly_Sales_Prediction seem reliable? How would you improve it?

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Actual vs Predicted
y_true = df['Last_Month_Spend']
y_pred = df['Monthly_Sales_Prediction']

# Metrics
MAE = mean_absolute_error(y_true, y_pred)
RMSE = np.sqrt(mean_squared_error(y_true, y_pred))
R2 = r2_score(y_true, y_pred)

MAE, RMSE, R2
```

Why we use all three:

MAE = average everyday error

RMSE = penalizes large mistakes

R² = overall goodness of fit

🔹 Interpretation

MAE ≈ 20,463
→ On average, the prediction is off by ₹20,000+ per customer/month.
That’s a very high error considering most monthly spends in your dataset are in the ₹20,000–30,000 range.

RMSE ≈ 22,005
→ Even worse, some predictions are hugely wrong (since RMSE > MAE).
This suggests there are big outliers where predictions miss by an extreme margin.

R² ≈ -8.02
→ A negative R² means the model is worse than simply predicting the average sales for everyone.
Basically, the prediction model fails to capture any meaningful relationship.

🔹 Conclusion

👉 The existing Monthly_Sales_Prediction is not reliable at all.


The current predictions are not useful; we need a better model with more features + proper error handling
